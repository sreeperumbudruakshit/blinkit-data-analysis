### DATA CLEANING


```python
print(df["Item Fat Content"].unique( ))
```

```python
df.replace(
    {"Item Fat Content": {
        "LF": "Low Fat",
        "low fat": "Low Fat",
        "reg": "Regular"
    }},
    inplace=True
)
```

### BUSINESS REQUIREMENTS

##### KPI REQUIREMENTS


```python
# Total sales
total_sales=df["Sales"].sum()
print(f"Total sales : ${total_sales:,.1f}")

# Average sales
avg_sales=df["Sales"].mean()
print(f"\nAverage sales : ${avg_sales:,.1f}")

# Sales by item type
sales_by_itemtype=df.groupby("Item Type")["Sales"].sum()
print(f"\nSales by item type : ${sales_by_itemtype}")

# no of items sold
no_of_items_sold=df["Sales"].count()
print(f"\nNo of items sold : {no_of_items_sold:,.1f}")

# Average ratings
avg_rating=df["Rating"].mean()
print(f"\nAverage rating : {avg_rating:,.1f}")

# Average rating by item type
avg_rating_by_itemtype=df.groupby("Item Type")["Rating"].mean().astype(int)
print(f"\nAverage rating by itemtype : {avg_rating_by_itemtype}")
```

### CHARTS REQUIREMENTS

#### Total sales by fat content

```python
sales_by_fat=df.groupby("Item Fat Content")["Sales"].sum()
print(sales_by_fat)
plt.pie(sales_by_fat,labels=sales_by_fat.index,autopct="%.1f%%",startangle=90)
plt.title("sales by fat content")
plt.axis("equal")
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/sales_by_fat.png", bbox_inches="tight")
plt.show()
```

##### Conclusion:
Low Fat products dominate total sales, contributing nearly two-thirds of overall sales, while Regular products contribute about one-third. This clearly shows a strong customer preference for low-fat and healthier food options.

#### TOTAL SALES BY ITEM TYPE

```python
sales_by_type=df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
bars=plt.bar(sales_by_type.index,sales_by_type.values)

plt.xticks(rotation=-90)
plt.xlabel("ITEM TYPE")
plt.ylabel("TOTAL SALES")
plt.title("TOTAL SALES BY ITEM TYPE")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
        f'{bar.get_height():,.0f}',ha='center',va='bottom',fontsize=8)
plt.tight_layout()
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/sales_by_type.png", bbox_inches="tight")
plt.show()
    
```

##### Conclusion:
Fruits and Vegetables and Snack Foods generate the highest sales among all item types.
Items like Seafood and Breakfast products have the lowest sales. This indicates that customers prefer everyday essential and snack items over specialty products.

#### FAT CONTENT BY OUTLET FOR TOTAL SALES

```python
fat_by_outlet=df.groupby(["Outlet Location Type", "Item Fat Content"])["Sales"].sum().unstack()
fat_by_outlet=fat_by_outlet[['Regular','Low Fat']]

ax=fat_by_outlet.plot(kind='bar',figsize=(8,5),title='Outlet Tier by Item Fat Content')
plt.xlabel("Outlet Location Tier")
plt.ylabel("Total Sales")
plt.title("Fat by content")
plt.tight_layout()
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/fat_by_outlet.png", bbox_inches="tight")
plt.show()
```

##### Conclusion:
In all outlet locations, Low Fat products sell more than Regular products.
Sales of both Low Fat and Regular products are highest in Tier 3 locations, showing strong demand for healthier options in these areas.

#### Total Sales by Outlet Establishment

```python
sales_by_year=df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index,sales_by_year.values,marker='o',linestyle='-')
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")
plt.title("Outlet Establishment")

for x,y in zip(sales_by_year.index,sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}',ha='center',va='bottom',fontsize=8)

plt.tight_layout()
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/sales_by_year.png", bbox_inches="tight")
plt.show()

```

##### Conclusion:
Older outlets (established around 1998) show very high sales, but sales are not consistently increasing with outlet age. Some newer outlets also perform well, indicating that outlet age alone does not guarantee higher sales.

#### SALES BY OUTLET SIZE

```python
Sales_by_size=df.groupby("Outlet Size")["Sales"].sum()

plt.pie(Sales_by_size,labels=Sales_by_size.index,autopct="%1.1f%%",startangle=90)
plt.title("Outlet Size")
plt.tight_layout()
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/sales_by_size.png", bbox_inches="tight")
plt.show()
```

##### Conclusion:
Medium-sized outlets generate the highest sales, followed by small outlets.
Large outlets contribute the least to total sales.
This suggests that medium stores are the most efficient in terms of sales performance.

#### SALES BY OUTLET LOCATION

```python
sales_by_location=df.groupby("Outlet Location Type")["Sales"].sum()
plt.pie(sales_by_location,labels=sales_by_location.index,autopct="%.1f%%",startangle=90)
plt.title("Outlet Location")
plt.tight_layout()
plt.savefig("C:/Users/akshi/Downloads/datascience/data cleaning/blinkit-data-analysis/images/sales_by_location.png", bbox_inches="tight")
plt.show()
```

##### Conclusion:
Most of the sales come from Tier 3 locations, followed by Tier 2, while Tier 1 locations contribute the least. This shows that outlets in smaller cities or towns are generating more sales than those in big cities.
